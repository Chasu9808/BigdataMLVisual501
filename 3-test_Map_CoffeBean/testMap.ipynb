```python
from folium import Choropleth
# 필요한 패키지 임포트
import numpy as np
import pandas as pd
import json
import folium
import geopandas as gpd
from geopandas import GeoDataFrame
from shapely.geometry import Polygon, LineString, Point
from fiona.crs import from_string

# 좌표 바꾸는 패키지
from pyproj import Transformer
import pyproj

library = pd.read_csv('./Library.csv')
seoul_boundary = gpd.read_file('./서울시구경계.shp', encoding='cp949')

# prepare and read data
F1 = './sig.shp' 
D1 = gpd.read_file(F1, encoding='cp949')
D1.head()

seoul_boundary.rename(columns={'SIGUNGU_NM':'구명'}, inplace=True)

seoul_boundary.rename(columns={'SIGUNGU_NM':'구명'}, inplace=True)

boundary = seoul_boundary[['구명', 'geometry']].set_index('구명')

library_count = library.구명.value_counts()

# Create a base map
m = folium.Map(location=[37.58, 127.0], tiles='cartodbpositron', zoom_start=11)

# Add a choropleth map
Choropleth(
    geo_data=boundary['geometry'].geometry.to_crs(epsg=4326).__geo_interface__,
    data=library_count,
    key_on='feature.id',
    fill_color='YlGnBu',
    legend_name='Seoul Libraries',
).add_to(m)

# Display the map
m
```
